*** Merging and Lookup Files (Master File Creation)***

```python
#86.Create a customer lookup DataFrame with unique CustomerID, CustomerName, and total orders. Save it.

customer_summary = data.groupby(['CustomerID', 'CustName']).agg(TotalOrders=('Product', 'count')).reset_index()
customer_summary.to_csv("Customer_Lookup.csv", index=False)
```

```python
#87.Create a product lookup DataFrame with unique Product, Category, and average Price.


product_summary = data.groupby(['Product', 'Category']).agg(AvgPrice=('Price', 'mean')).reset_index()
product_summary.to_csv("Product_Lookup.csv", index=False)
```

```python
#88.Merge the original Sales_100k_Dataset.csv with both the customer and product lookup


data = data.merge(customer_summary, on=['CustomerID', 'CustName'], how='left')
data = data.merge(product_summary, on=['Product', 'Category'], how='left')
```
